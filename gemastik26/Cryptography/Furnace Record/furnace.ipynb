# Furnace Record

The spool was copied from the west furnace terminal. The old note says the wheel must be advanced exactly as written before the capsule opens.

Put `spool.bin` next to this notebook.

```python
import json, hashlib

raw = open('spool.bin', 'rb').read().splitlines()[1]
spool = json.loads(raw)
n = int(spool['n'], 16)
x = int(spool['a'], 16)

for _ in range(spool['turns']):
    x = (x * x) % n

seed = x.to_bytes((n.bit_length() + 7) // 8, 'big')
capsule = bytes.fromhex(spool['capsule'])
stream = b''
i = 0
while len(stream) < len(capsule):
    stream += hashlib.sha256(seed + i.to_bytes(4, 'big')).digest()
    i += 1

bytes(a ^ b for a, b in zip(capsule, stream))
```
